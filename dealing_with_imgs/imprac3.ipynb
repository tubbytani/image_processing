```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1=plt.imread(r"C:\Users\new\Documents\Python\MARB.tif")
r=img1[:,:,0]
g=img1[:,:,1]
b=img1[:,:,2]
```

```python
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.subplot(2,2,1)
plt.title("original image")
plt.imshow(img1)

hist,bins=np.histogram(r.ravel(),bins=256,range=(0,256)) #flattening is must
plt.subplot(2,2,2)
plt.title("red histogram")
plt.bar(bins[:-1],hist)

hist,bins=np.histogram(g.ravel(),bins=256,range=(0,256)) #flattening is must
plt.subplot(2,2,3)
plt.title("green histogram")
plt.bar(bins[:-1],hist)

hist,bins=np.histogram(b.ravel(),bins=256,range=(0,256)) #flattening is must
plt.subplot(2,2,4)
plt.title("blue histogram")
plt.bar(bins[:-1],hist)
plt.show()
```

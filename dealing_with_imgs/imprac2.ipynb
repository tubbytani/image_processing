```python
imgg1=plt.imread(r"C:\Users\new\Documents\Python\1.2.13.tiff")
imgg2=plt.imread(r"C:\Users\new\Documents\Python\1.2.12.tiff") 
#should be of same size or matrix should be 1 otherwise broadcasting error 
```

```python
plt.imshow(np.bitwise_and(imgg1,imgg2)) 
```

```python
plt.imshow(np.bitwise_or(imgg1,imgg2)) 
```

```python
plt.subplot(1,2,1)
plt.imshow(np.bitwise_xor(imgg1,imgg2)) 
plt.subplot(1,2,2)
plt.imshow(np.bitwise_and(imgg1,imgg2)) 
plt.show()
```

```python
plt.subplot(1,2,1)
plt.imshow(imgg1,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(np.bitwise_not(imgg1),cmap="gray") #negative of bw image
plt.show()
```

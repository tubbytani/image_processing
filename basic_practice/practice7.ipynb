```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4],dtype=np.uint8)
plt.title("single line")
plt.plot(x)
plt.show()
```

```python
plt.plot(x,[y**2 for y in x]) #parantheseis is necessary for y
plt.show()
```

```python
#multiline plots
```

```python
plt.plot(x,[y**2 for y in x]) #or can do plt.plot(x,y,x1,y1...)
plt.plot(x,2*x)
plt.plot(x,x*3)
plt.show()
```

```python
x=np.array([[1,2,3],[4,5,6]])
plt.plot(x)
plt.show()
```

```python
y=np.random.rand(3,5)
plt.plot(y[0],y[1])
plt.show
```

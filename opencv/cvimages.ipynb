```python
img1=cv2.imread(r"C:\Users\new\Documents\Python\1.2.12.tiff",1)
img2=cv2.imread(r"C:\Users\new\Documents\Python\1.2.13.tiff",1)
```

```python
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
multiplication=img1*img2 #img1 and img2 are actualy numpy arrays
division=img1/img2
```

```python
titles=["liquid drop","mandrill","multiplication" ,"division"]
```

```python
images=[img1 ,img2, multiplication, division]
```

```python
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])# Disable xticks.
    plt.yticks([])#disable yticks
plt.show()    
```

```python
#image blending
```

```python
coefA=0.5
coefB=0.5
gamma=0
```

```python
output=cv2.addWeighted(img1,coefA,img2,coefB,gamma)
titles2=["image1","image2","blended"]
images=[img1,img2,output]
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(titles2[i])
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
plt.show()    
```

```python
#image trasition
```

```python
for i in np.linspace(0,1,500):
    coefA=i
    coefB=1-coefA
    gamma=0
    output=cv2.addWeighted(img1,coefA,img2,coefB,gamma)
    cv2.imshow('transition',output)
    time.sleep(0.02)
    if cv2.waitKey(1)==27:#stops the execution of current thread only, not the whole program.
        break
cv2.destroyAllWindows()
```

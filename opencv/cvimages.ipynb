```python
rows,coloumns,channels=img1.shape
T=np.float32([[1,0,150],[0,1,50]])
print(T)
output=cv2.warpAffine(img1,T,(coloumns,rows))#T=[1 0 Tx]
plt.imshow(output)                            # [0 1 Ty]
plt.title('shifted image')                  #Tx and Ty are the tarnsformation spaces
```

```python
R=cv2.getRotationMatrix2D((coloumns/2,rows/2),50,2) #50-angle,2 is the zooming efect ie.enlarged if less than 1 then smaller
print(R)
output=cv2.warpAffine(img1,R,(coloumns,rows))
plt.imshow(output)
plt.title('rotated image')
```

```python
angle=360
k=1
while True:
    if angle==0:
        angle=360
    R=cv2.getRotationMatrix2D((coloumns/2,rows/2),angle,0.5)
   # if k==1:
   #     print(R)
   #     k=k+1
   # else:
   #     k=k+1
    output=cv2.warpAffine(img1,R,(coloumns,rows))  
    cv2.imshow('rotated image',output)
    angle=angle-1
    time.sleep(0.01)
    if cv2.waitKey(1)==27:
        break
cv2.destroyWindow('Rotated image')        
```

```python
winname="live video feed"
cv2.namedWindow("winname")
cap=cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame=cap.read()
else:
    ret =False
rows,coloumns,chanells=frame.shape
angle=0
scale=0.1
while True:    
    ret,frame=cap.read()
    angle=0
    if scale<2:
        scale=scale+0.1
    if scale>=2:
        scale=0.1
   # print(scale)    
    R=cv2.getRotationMatrix2D((coloumns/2,rows/2),angle,scale)
   # print(R)
    output=cv2.warpAffine(frame,R,(coloumns,rows))  
    time.sleep(0.01)
    cv2.imshow(winname,output)
    if cv2.waitKey(1)==27:
        break
cv2.destroyWindow(winname)
cv2.destroyAllWindows()
cap.release()    
```

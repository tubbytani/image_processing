```python
import cv2
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1.global thresholding: simple thresholding (better seen if used a gradient image)
```

```python
img1=cv2.imread(r"C:\Users\new\Documents\Python\gray band.tif",0)#0 to make it into grayscale
th=137
max_val=255#this value is set if greater than threshold value
ret,o1=cv2.threshold(img1,th,max_val,cv2.THRESH_BINARY)
ret,o2=cv2.threshold(img1,th,max_val,cv2.THRESH_BINARY_INV)
#print(ret) 137
ret,o3=cv2.threshold(img1,th,max_val,cv2.THRESH_TOZERO)
ret,o4=cv2.threshold(img1,th,max_val,cv2.THRESH_TOZERO_INV)
ret,o5=cv2.threshold(img1,th,max_val,cv2.THRESH_TRUNC)
output=[img1,o1,o2,o3,o4,o5]
title=["original",'binary','binary inv','zero','zero inverted','trunc']
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(title[i])
    plt.imshow(output[i],cmap='gray')
    plt.xticks([])
    plt.yticks([])
```

```python
img1=cv2.imread(r"C:\Users\new\Documents\Python\gray band.tif",0)#0 to make it into grayscale
th=0
max_val=255#this value is set if greater than threshold value
ret,o1=cv2.threshold(img1,th,max_val,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,o2=cv2.threshold(img1,th,max_val,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
#print(ret) 137
ret,o3=cv2.threshold(img1,th,max_val,cv2.THRESH_TOZERO+cv2.THRESH_OTSU)
ret,o4=cv2.threshold(img1,th,max_val,cv2.THRESH_TOZERO_INV+cv2.THRESH_OTSU)
ret,o5=cv2.threshold(img1,th,max_val,cv2.THRESH_TRUNC+cv2.THRESH_OTSU)
output=[img1,o1,o2,o3,o4,o5]
title=["original",'binary','binary inv','zero','zero inverted','trunc']
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(title[i])
    plt.imshow(output[i],cmap='gray')
    plt.xticks([])
    plt.yticks([])
```

```python
#b. local thresholding :adaptive thresholding
```

```python
img1=cv2.imread(r"C:\Users\new\Documents\Python\MARB.tif",0)#0 to make it into grayscale
block_size=23#vary it to see the difference
constant=2
th1=cv2.adaptiveThreshold(img1,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,block_size,constant)
th2=cv2.adaptiveThreshold(img1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,block_size,constant)
output=[img1,th1,th2]
title=["original",'mean adpt.binary','gaussian adpt.binary inv']
for i in range(3):
    plt.subplot(2,2,i+1)
    plt.title(title[i])
    plt.imshow(output[i],cmap='gray')
    plt.xticks([])
    plt.yticks([])
```

```python
img3=np.zeros((512,512,3),dtype=np.uint8)
#img3 = np.array([512,512,3]).astype(np.uint8)
#cv2.resize(img3,dsize=(512,512) ,interpolation=cv2.INTER_LINEAR)
cv2.line(img3,(0,99),(99,0),(255,0,0),2)#coordinates color thickness
cv2.rectangle(img3,(100,60),(200,170),(0,255,0),2) #vertices color thickness
cv2.ellipse(img3,(160,260),(50,20),0,0,360,(127,127,127),-1)# major, minor axis color fill area
cv2.circle(img3,(60,60),50,(0,0,255),-1)#center radius color fill the entire shape with the given color
#polygon
points=np.array([[80,20],[125,0],[179,0],[230,5],[30,50]],np.int32)
points=points.reshape(-1,1,2)
cv2.polylines(img3,points,True,(0,255,255))
text1="text"
cv2.putText(img3,text1,(100,100),cv2.FONT_HERSHEY_SIMPLEX,5,(255,255,0))#position font size color
cv2.imshow("shapes",img3)
cv2.waitKey(0)
cv2.destroyWindow("shapes")
```

```python
def emptyfunc(self):
      pass
```

```python
img4=np.zeros((512,512,3),dtype=np.uint8)
name="color palette"
cv2.namedWindow(name)

cv2.createTrackbar('B',name,0,255,emptyfunc) #trackbar name window name 0-255 range
cv2.createTrackbar('R',name,0,255,emptyfunc)
cv2.createTrackbar('G',name,0,255,emptyfunc)
while (True):
    cv2.imshow(name,img4)
    
    if cv2.waitKey(1)==27: #it works with esc key only
        break    
    blue=cv2.getTrackbarPos('B',name)
    red=cv2.getTrackbarPos('R',name)
    green=cv2.getTrackbarPos('G',name)
    img4[:]=[blue,green,red] #in the loop
cv2.destroyAllWindows() 
```

```python
name="drawing app"
img=np.zeros((512,512,3),np.uint8) #black canvas
cv2.namedWindow(name)

def draw_circle(event,x,y,flags,param):#events and binds(to the events)
    if event==cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img,(x,y),40,(0,255,0),-1) #after 40 center rest are parameters
    if event==cv2.EVENT_MBUTTONDOWN: 
        cv2.circle(img,(x,y),20,(0,255,255),-1)
    if event==cv2.EVENT_LBUTTONDOWN:    
        cv2.circle(img,(x,y),30,(255,0,0),-1)

#bind the callback function to the window
cv2.setMouseCallback(name,draw_circle)
while (True):
    cv2.imshow(name,img)
    if cv2.waitKey(1)==27: #it works with esc key only ascii ==27
        break 

cv2.destroyAllWindows()
```

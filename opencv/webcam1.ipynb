```python
name="vid_feed"
cv2.namedWindow(name)
cap=cv2.VideoCapture(0)

if cap.isOpened:
    ret,frame=cap.read()
else:
    ret=False
while ret:
    ret,frame=cap.read()
    ouput=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow(name,frame)
    if cv2.waitKey(1)==27:
        break
cv2.destroyAllWindows()
cap.release()
```
